# Run EddyMotionEstimator with chunk-by-chunk parallelization of tensormodel fit

## Imports

```python
from pathlib import Path
from eddymotion import dmri
from eddymotion.estimator import EddyMotionEstimator
import numpy as np
import nitransforms as nt
from dipy.io.image import load_nifti
import matplotlib.pyplot as plt
```

## Paths setup

```python
# Set the installation directory and cmp derivatives directory where DTI and mask are located.
root_dir = Path('exam_data')

dwi_file = root_dir/'sub-01_desc-preproc_dwi.nii.gz'
bvec_file = root_dir/'sub-01_desc-cmp_dwi.bvec'
bval_file = root_dir/'sub-01_desc-cmp_dwi.bval'
brain_mask_file = root_dir/'sub-01_desc-brain_mask.nii.gz'
mean_B0_file = root_dir/'sub-01_dwi_desc-mean_b0.nii.gz'
rasb_file = root_dir/'sub-01_dwi_rasb.tsv'
```

## Define functions and create a reference B0 and the gradient in RAS+B format

```python
data = dmri.load(
    str(dwi_file),
    gradients_file=str(rasb_file),
    b0_file=str(mean_B0_file),
    brainmask_file=str(brain_mask_file)
    )
```

```python
em_affines = EddyMotionEstimator.fit(data, model="b0")
```

```python
np.save("result.npy", em_affines)
```

```python
res = np.load("result.npy",allow_pickle=True)
```

```python
print(res)
```

```python
data, affine, img = load_nifti(dwi_file, return_img=True) 
```

```python
axial_middle = data.shape[2] // 2
plt.figure()
for i in range(3):
    plt.subplot(1, 3, i+1).set_axis_off()
    plt.imshow(data[:, :, axial_middle, i].T, cmap='gray', origin='lower')
plt.show()
```

```python
data.em_affines = em_affines
```

```python
data.to_nifti("realigned.nii.gz")
```

```python
data, affine, img = load_nifti("realigned.nii.gz", return_img=True) 
```

```python
axial_middle = data.shape[2] // 2
plt.figure()
for i in range(3):
    plt.subplot(1, 3, i+1).set_axis_off()
    plt.imshow(data[:, :, axial_middle, i].T, cmap='gray', origin='lower')
eddymotion_correction_pic = "eddymotion_correction.png"
plt.savefig(eddymotion_correction_pic, bbox_inches='tight')
plt.show()
```
